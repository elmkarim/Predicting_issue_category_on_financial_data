```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from helper_save_load import load_from_pickle

df_a, df_f, df_v = load_from_pickle("data/dataframes_Dollars.pickle")

df_c = load_from_pickle("data/Labeled_comments.pkl")

ignored_brands = ["57-CAMAY BAR SOAP Brand",
"IGNORE",
"71-LBP SH/CD Roll-up Brand",
"72-LBP SHOWER Roll-up Brand"]

df_c = df_c[~df_c["Brand_1"].isin(ignored_brands)]
```

```python
class_label_cols = ["Phasing", "Promo", "SP&D", "POS"]
territory_label_cols = df_c.columns[17:-4]
```

```python
class_labels = df_c[class_label_cols]
territory_labels = df_c[territory_label_cols]
```

```python
# Unlabeled commentaries
df_c[df_c["Commentaries"].notnull()][class_label_cols].isnull().apply(lambda x: all(x), axis=1).sum()
```

```python
class_labels.sum().sort_values(ascending=False).plot(kind='bar', color='r')
plt.title("Class Frequency", fontsize=14)
plt.show()

plt.figure(figsize=(10, 4))
territory_labels.sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Territory Frequency", fontsize=14)
plt.show()
```

```python
def group_by_TR_BR(territory, brand):
    
    v = df_v.groupby(["Territory", "Brand"]).sum().loc[territory].loc[brand]
    a = df_a.groupby(["Territory", "Brand"]).sum().loc[territory].loc[brand]
    f = df_f.groupby(["Territory", "Brand"]).sum().loc[territory].loc[brand]
    
    a.index = v.index
    f.index = v.index

    return v, a, f
```

```python
def plot_actualforecast(actual, forecast):
    plt.figure(figsize=(30,10))
    plt.plot(actual, label="Actual")
    plt.plot(forecast, label="Forecast")
    plt.xticks(rotation=90, fontsize=18)
    plt.yticks(fontsize=18)
    plt.title(actual.name, fontsize=24)
    plt.grid(axis='x')
    plt.xlim(0, len(actual)-1)

    plt.legend(fontsize=18)
    plt.show()
```

```python
def plot_variance(variance):
    plt.figure(figsize=(30,10))
    plt.plot(variance)

    plt.xticks(rotation=90, fontsize=18)
    plt.yticks(fontsize=18)
    plt.title(variance.name, fontsize=24)
    plt.grid(axis='x')
    plt.hlines(0, 0, len(variance)-1)
    plt.xlim(0, len(variance)-1)

    plt.show()
```

```python
v, a, f = group_by_TR_BR("M7500 - CANADA COSTCO", "MY-Q-TIPS SWABS Brand")
plot_actualforecast(a, f)
plot_variance(v)
```
